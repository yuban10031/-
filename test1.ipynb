```python
commentURL='http://comment5.news.sina.com.cn/page/info?version=1&format=js&\
channel=fj&newsid=comos-{}&group=&compress=0&ie=u\
tf-8&oe=utf-8&page=1&page_size=20'
def getCommentCount(newsurl):
    m = re.search('detail-(.+).shtml',newsurl)
    newsid=m.group(1)
    comments = requests.get(commentURL.format(newsid))
    jd=json.loads(comments.text.strip('var data='))
    return jd['result']['count']['total']
```

```python
import re
import json
def getCommentCounts(newsurl):
    m = re.search('doc-i(.+).shtml',newsurl)
    newsid = m.group(1)
    comments = requests.get(commentURL.format(newsid))
    jd=json.loads(comments.text.strip('var data='))
    return jd['result']['count']['total']
```

```python
import requests
from bs4 import BeautifulSoup
def getNwesDetail(newsurl):
    result={ }
    res=requests.get(newsurl)
    res.encoding = 'utf - 8'
    soup = BeautifulSoup(res.text,'html.parser')
    result['title'] =soup.select('#artibodyTitle')[0].text 
    result['newssource']=soup.select('.time-source span a')[0].text
    result['timesource']=soup.select('.time-source')[0].contents[0].strip()
    return result
```
